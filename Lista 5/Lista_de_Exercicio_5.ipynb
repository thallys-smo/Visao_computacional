### 1) Momentos (Nota: 1.5/10)

A equação generalizada para o cálculo do momento $m_{pq}$ é dada por:

\begin{equation}
m_{pq} = \int_{-\infty}^{\infty}\int_{-\infty}^{\infty} x^p y^q \, f(x,y) \, dx dy, \end{equation}

onde $f(x,y)$ é a imagem, $x$ e $y$ suas respectivas coordenadas. Seguindo a teoria de momentos, as coordenadas do centro de massa normalizado de um objeto podem ser determinadas utilizando as seguintes equações:

\begin{equation}
  \bar{x} = \frac{m_{10}}{m_{00}}
  \;\;\;
  \bar{y} = \frac{m_{01}}{m_{00}}
\end{equation}

````python
def momentos(image):

  momento_00,momento_10,momento_01,momento_11 = 0, 0, 0, 0 #Atribuindo 0 aos momentos  
  
  for x in range(image.shape[0]):
    for y in range(image.shape[1]):
      momento_00 = momento_00 + image[x,y]     #Cálculo de M00 (p = 0 e q = 0)
      momento_10 = momento_10 + x*image[x,y]   #Cálculo de M10 (p = 1 e q = 0)
      momento_01 = momento_01 + y*image[x,y]   #Cálculo de M01 (p = 0 e q = 1)
      momento_11 = momento_11 + x*y*image[x,y] #Cálculo de M11 (p = 1 e q = 1)

  return momento_00, momento_01, momento_10, momento_11

#Importar imagens do triangulo e da estrela em RGB
img_tri = cv.imread("triangulo.png")
img_est = cv.imread("estrela.png")

#Converter imagens do triângulo e da estrela em gray scale
img_tri_g = cv.cvtColor(img_tri, cv.COLOR_BGR2GRAY)
img_est_g = cv.cvtColor(img_est, cv.COLOR_BGR2GRAY)

#Momentos m00, m10, m01 e m11 das imagens
momentos_img = []
momentos_img.append(momentos(img_tri_g))
momentos_img.append(momentos(img_est_g))

print("Momentos - Imagem Triângulo:")
print("  Momento 00: {}".format(momentos_img[0][0]))
print("  Momento 01: {}".format(momentos_img[0][1]))
print("  Momento 10: {}".format(momentos_img[0][2]))
print("  Momento 11: {}".format(momentos_img[0][3]))

print("\n")

print("Momentos - Imagem Estrela:")
print("  Momento 00: {}".format(momentos_img[1][0]))
print("  Momento 01: {}".format(momentos_img[1][1]))
print("  Momento 10: {}".format(momentos_img[1][2]))
print("  Momento 11: {}".format(momentos_img[1][3]))

# Centro de Massas das Imagens

x_tri = momentos_img[0][2]/momentos_img[0][0]
y_tri = momentos_img[0][1]/momentos_img[0][0]
x_est = momentos_img[1][2]/momentos_img[1][0]
y_est = momentos_img[1][1]/momentos_img[1][0]

#Printar e desenhar os centróides calculados (através das relaçõs m10/m00 e m01/m00)

print("Centroide triângulo = (", x_tri, ", ",y_tri,")")
print("Centroide estrela = (", x_est, ", ",x_est,")")
print("\n")

#Resultados
plt.figure(figsize = (15,15))
plt.subplot(2,2,1)
plt.imshow(img_tri_g,cmap = "gray")
plt.title("Imagem Triângulo (Escala de Cinza) - Não Rotacionado")

plt.subplot(2,2,2)
plt.imshow(img_est_g,cmap = "gray")
plt.title("Imagem Estrela Original (Escala de Cinza) - Não Rotacionado")

plt.subplot(2,2,3)
plt.imshow(img_tri)
plt.plot(x_tri,y_tri,"+")
plt.title("Imagem Triângulo com Centróide - Não Rotacionado")

plt.subplot(2,2,4)
plt.imshow(img_est)
plt.plot(x_est,y_est,"+")
plt.title("Imagem Estrela com Centróide - Não Rotacionado")

# Rotacionado as Imagems em 180

img_tri_rot = cv.rotate(img_tri, cv.ROTATE_180)
img_est_rot = cv.rotate(img_est, cv.ROTATE_180)

#Converter imagens do triângulo e da estrela em gray scale
img_tri_rot_g = cv.cvtColor(img_tri_rot, cv.COLOR_BGR2GRAY)
img_est_rot_g = cv.cvtColor(img_est_rot, cv.COLOR_BGR2GRAY)

#Momentos m00, m10, m01 e m11 das imagens
momentos_img_rot = []
momentos_img_rot.append(momentos(img_tri_rot_g))
momentos_img_rot.append(momentos(img_est_rot_g))

print("Momentos - Imagem Triângulo Rotacionado :")
print("  Momento 00: {}".format(momentos_img_rot[0][0]))
print("  Momento 01: {}".format(momentos_img_rot[0][1]))
print("  Momento 10: {}".format(momentos_img_rot[0][2]))
print("  Momento 11: {}".format(momentos_img_rot[0][3]))

print("\n")

print("Momentos - Imagem Estrela Rotacionado :")
print("  Momento 00: {}".format(momentos_img_rot[1][0]))
print("  Momento 01: {}".format(momentos_img_rot[1][1]))
print("  Momento 10: {}".format(momentos_img_rot[1][2]))
print("  Momento 11: {}".format(momentos_img_rot[1][3]))
print("\n")


# Centro de Massas das Imagens

x_tri = momentos_img_rot[0][2]/momentos_img_rot[0][0]
y_tri = momentos_img_rot[0][1]/momentos_img_rot[0][0]
x_est = momentos_img_rot[1][2]/momentos_img_rot[1][0]
y_est = momentos_img_rot[1][1]/momentos_img_rot[1][0]

print("Centroide Triângulo Rotacionado = (", x_tri, ", ",y_tri,")")
print("Centroide Estrela Rotacionado = (", x_est, ", ",x_est,")")

print("\n")

#Resultados
plt.figure(figsize = (15,15))
plt.subplot(2,2,1)
plt.imshow(img_tri_rot_g,cmap = "gray")
plt.title("Imagem Triângulo (Escala de Cinza) - Rotacionada")

plt.subplot(2,2,2)
plt.imshow(img_est_rot_g,cmap = "gray")
plt.title("Imagem Estrela Original (Escala de Cinza) - Rotacionada")

plt.subplot(2,2,3)
plt.imshow(img_tri_rot)
plt.plot(x_tri,y_tri,"+")
plt.title("Imagem Triângulo com Centróide - Rotacionada")

plt.subplot(2,2,4)
plt.imshow(img_est_rot)
plt.plot(x_est,y_est,"+")
plt.title("Imagem Estrela com Centróide - Rotacionada")
````

````python
#Importar imagens em grayscale
imgGuitarra = cv.imread("guitarra.tif",0)
imgBanjo = cv.imread("banjo.tif",0)

#Binarização da imagem utilizando um determinado valor de limiar (encontrado manualmente)
retGuitarra,thrGuitarra = cv.threshold(imgGuitarra,245,255,cv.THRESH_BINARY_INV)  #binarização da guitarra
retBanjo,thrBanjo = cv.threshold(imgBanjo,245,255,cv.THRESH_BINARY_INV)  #binarização da guitarra

#Criação de uma imagem do tamanho da guitarra
imgBlackG = np.zeros([imgGuitarra.shape[0],imgGuitarra.shape[1]],np.uint8)
imgBlackB = np.zeros([imgBanjo.shape[0],imgBanjo.shape[1]],np.uint8)

#Redução da imagem da guitarra e do banjo utilizando a função resize e inserir ela na imagem de fundo do tamanho original das imagens
hg = imgGuitarra.shape[0]
wg = imgGuitarra.shape[1]
hb = imgBanjo.shape[0]
wb = imgBanjo.shape[1]
guitarraResized = cv.resize(thrGuitarra, (int(hg/4),int(wg/4)))  
banjoResized = cv.resize(thrBanjo, (int(hb/4),int(wb/4)))

imgBlackG[180:guitarraResized.shape[0]+180, 200:guitarraResized.shape[1]+200] = guitarraResized
imgBlackB[180:banjoResized.shape[0]+180, 200:banjoResized.shape[1]+200] = banjoResized

#Rotação da imagens
imgRotatedG = rotate(imgBlackG, 30, preserve_range=True)
imgRotatedB = rotate(imgBlackB, 30, preserve_range=True)

#Cálculo dos momentos
momentsGuitarra = cv.moments(thrGuitarra)             #momentos da imagem da guitarra sem transfomração
momentsBanjo = cv.moments(thrBanjo)                   #momentos da imagem do banjo sem transformação
momentsGuitarraTransformed = cv.moments(imgRotatedG)  #momentos da imagem da guitarra transformada
momentsBanjoTransformed = cv.moments(imgRotatedB)     #momentos da imagem do banjo transformada


#Cálculo dos momentos invariantes de HU das imagens
HuMomentsGuitarra = cv.HuMoments(momentsGuitarra)                         #momentos invariantes de HU da imagem da guitarra sem transformação
HuMomentsBanjo = cv.HuMoments(momentsBanjo)                               #momentos invariantes de HU da imagem do banjo sem transformação
HuMomentsGuitarraTransformed = cv.HuMoments(momentsGuitarraTransformed)   #momentos invariantes de HU da imagem da guitarra transformada
HuMomentsBanjoTransformed = cv.HuMoments(momentsBanjoTransformed)         #momentos invariantes de HU da imagem do banjo transformada


#Comparação dos momentos invariantes de HU
print("Imagem da Guitarra original: ",HuMomentsGuitarra)
print("Imagem da Guitarra transformada: ",HuMomentsGuitarraTransformed)
print("\n")
print("Imagem do Banjo original: ",HuMomentsBanjo)
print("Imagem do Banjo transformada: ",HuMomentsBanjoTransformed)
print("\n")
print("Imagem da Guitarra original: ",HuMomentsGuitarra)
print("Imagem do Banjo original: ",HuMomentsBanjo)
print("\n")
print("Imagem da Guitarra transformada: ",HuMomentsGuitarraTransformed)
print("Imagem do Banjo transformada: ",HuMomentsBanjoTransformed)
print("\n")

#Distância euclidiana
print("Distância euclidiana entre a guitarra original e guitarra transformada: ",np.linalg.norm(HuMomentsGuitarra - HuMomentsGuitarraTransformed))
print("Distância euclidiana entre o banjo original e banjo transformado: ",np.linalg.norm(HuMomentsBanjo - HuMomentsBanjoTransformed))
print("Distância euclidiana entre o banjo original e guitarra transformada: ",np.linalg.norm(HuMomentsBanjo - HuMomentsGuitarraTransformed))
print("Distância euclidiana entre o banjo transformado e guitarra original: ",np.linalg.norm(HuMomentsBanjoTransformed - HuMomentsGuitarra))
print("Distância euclidiana entre o banjo transformado e guitarra transformada: ",np.linalg.norm(HuMomentsBanjoTransformed - HuMomentsGuitarraTransformed))
print("Distância euclidiana entre o banjo original e guitarra original: ",np.linalg.norm(HuMomentsBanjo - HuMomentsGuitarra))
print("\n")


#Resultados
plt.figure(figsize = (12,12))
plt.subplot(3,2,1)
plt.imshow(imgGuitarra, cmap = "gray")
plt.title("Imagem da guitarra original")

plt.subplot(3,2,2)
plt.imshow(imgBanjo, cmap = "gray")
plt.title("Imagem do banjo original")

plt.subplot(3,2,3)
plt.imshow(thrGuitarra, cmap = "gray")
plt.title("Binarização da guitarra")

plt.subplot(3,2,4)
plt.imshow(thrBanjo, cmap = "gray")
plt.title("Binarização da Banjo")

plt.subplot(3,2,5)
plt.imshow(imgRotatedG, cmap = "gray")
plt.title("Imagem da guitarra transformada")

plt.subplot(3,2,6)
plt.imshow(imgRotatedB, cmap = "gray")
plt.title("Imagem do banjo transformada")


plt.show()
````

````python
def assinatura(img, ic, jc):
  '''
  Descrição: Cria assinatura para uma imagem.

  Entrada:
    - img: Imagem de entrada (uint8).
    - ic: Coordenada X do centro de massa normalizado (uint8).
    - jc: Coordenada Y do centro de massa normalizado (uint8).

  Saída:
    - rho: Coordenadas polar (rho) dos pontos [range=0:img.shape[0]*np.sqrt(2)/2].
    - theta: Coordenadas polar (theta) dos pontos [range=-180:180].
  '''
  # Sistema de coordenadas
  # .__> jj (y)
  # |
  # v
  # ii (x)

  # Teste de insanidade
  if img.ndim != 2:
    raise ValueError('Imagem deve ter 2 dimensões.')

  # Varre a imagem, determinando as coordenadas dos pontos brancos

  contours, hierarchy = cv.findContours(image=img, mode=cv.RETR_TREE, method=cv.CHAIN_APPROX_NONE)

  # Translada a origem do sistema de coordenadas para o centro de massa

  img_copy = np.zeros(img.shape)

  cv.drawContours(image=img_copy, contours=contours, contourIdx=1, color=(255, 255, 255), thickness=1, lineType=cv.LINE_AA)


  ii,jj = np.nonzero(img_copy)
  ii -= np.int(ic)
  jj -= np.int(jc)
  # Converte as novas coordenadas cartesianas para coordenadas polares
  rho = np.sqrt(ii**2 + jj**2)
  theta = np.rad2deg(np.arctan2(jj, ii))

  return rho, theta

#Importar as imagens em grayscale
img1 = cv.imread("circulo1.tif",0)
img2 = cv.imread("circulo2.tif",0)
img3 = cv.imread("quadrado1.tif",0)
img4 = cv.imread("quadrado2.tif",0)

#Função para o cálculo dos momentos
(m00_1,m10_1,m01_1, m11_1) = momentos(img1)
(m00_2,m10_2,m01_2, m11_2) = momentos(img2)
(m00_3,m10_3,m01_3, m11_3) = momentos(img3)
(m00_4,m10_4,m01_4, m11_4) = momentos(img4)

#Cálculo dos centróides (xi,yi)
x1 = m10_1/m00_1
y1 = m01_1/m00_1

x2 = m10_2/m00_2
y2 = m01_2/m00_2

x3 = m10_3/m00_3
y3 = m01_3/m00_3

x4 = m10_4/m00_4
y4 = m01_4/m00_4


#Aplicação da função assinatura
rho1, theta1 = assinatura(img1,x1,y1)
rho2, theta2 = assinatura(img2,x2,y2)
rho3, theta3 = assinatura(img3,x3,y3)
rho4, theta4 = assinatura(img4,x4,y4)

#Resultados
plt.figure(figsize = (15,15))
plt.subplot(4,2,1)
plt.imshow(img1,cmap = 'gray')
plt.plot(x1,y1,'+')
plt.title("circulo1")

plt.subplot(4,2,2)
plt.plot(theta1,rho1,'.')
plt.title("assinatura circulo1")

plt.subplot(4,2,3)
plt.imshow(img2,cmap = 'gray')
plt.plot(x2,y2,'+')
plt.title("circulo2")

plt.subplot(4,2,4)
plt.plot(theta2,rho2,'.')
plt.title("assinatura circulo2")

plt.subplot(4,2,5)
plt.imshow(img3,cmap = 'gray')
plt.plot(x3,y3,'+')
plt.title("quadrado1")

plt.subplot(4,2,6)
plt.plot(theta3,rho3,'.')
plt.title("assinatura quadrado1")

plt.subplot(4,2,7)
plt.imshow(img4,cmap = 'gray')
plt.plot(x4,y4,'+')
plt.title("quadrado2")

plt.subplot(4,2,8)
plt.plot(theta4,rho4,'.')
plt.title("assinatura quadrado2")

plt.show()


````
