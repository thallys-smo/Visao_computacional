### 1) Momentos (Nota: 1.5/10)

A equação generalizada para o cálculo do momento $m_{pq}$ é dada por:

\begin{equation}
m_{pq} = \int_{-\infty}^{\infty}\int_{-\infty}^{\infty} x^p y^q \, f(x,y) \, dx dy, \end{equation}

onde $f(x,y)$ é a imagem, $x$ e $y$ suas respectivas coordenadas. Seguindo a teoria de momentos, as coordenadas do centro de massa normalizado de um objeto podem ser determinadas utilizando as seguintes equações:

\begin{equation}
  \bar{x} = \frac{m_{10}}{m_{00}}
  \;\;\;
  \bar{y} = \frac{m_{01}}{m_{00}}
\end{equation}

````python
def momentos(image):

  momento_00,momento_10,momento_01,momento_11 = 0, 0, 0, 0 #Atribuindo 0 aos momentos  
  
  for x in range(image.shape[0]):
    for y in range(image.shape[1]):
      momento_00 = momento_00 + image[x,y]     #Cálculo de M00 (p = 0 e q = 0)
      momento_10 = momento_10 + x*image[x,y]   #Cálculo de M10 (p = 1 e q = 0)
      momento_01 = momento_01 + y*image[x,y]   #Cálculo de M01 (p = 0 e q = 1)
      momento_11 = momento_11 + x*y*image[x,y] #Cálculo de M11 (p = 1 e q = 1)

  return momento_00, momento_01, momento_10, momento_11

#Importar imagens do triangulo e da estrela em RGB
img_tri = cv.imread("triangulo.png")
img_est = cv.imread("estrela.png")

#Converter imagens do triângulo e da estrela em gray scale
img_tri_g = cv.cvtColor(img_tri, cv.COLOR_BGR2GRAY)
img_est_g = cv.cvtColor(img_est, cv.COLOR_BGR2GRAY)

#Momentos m00, m10, m01 e m11 das imagens
momentos_img = []
momentos_img.append(momentos(img_tri_g))
momentos_img.append(momentos(img_est_g))

print("Momentos - Imagem Triângulo:")
print("  Momento 00: {}".format(momentos_img[0][0]))
print("  Momento 01: {}".format(momentos_img[0][1]))
print("  Momento 10: {}".format(momentos_img[0][2]))
print("  Momento 11: {}".format(momentos_img[0][3]))

print("\n")

print("Momentos - Imagem Estrela:")
print("  Momento 00: {}".format(momentos_img[1][0]))
print("  Momento 01: {}".format(momentos_img[1][1]))
print("  Momento 10: {}".format(momentos_img[1][2]))
print("  Momento 11: {}".format(momentos_img[1][3]))

# Centro de Massas das Imagens

x_tri = momentos_img[0][2]/momentos_img[0][0]
y_tri = momentos_img[0][1]/momentos_img[0][0]
x_est = momentos_img[1][2]/momentos_img[1][0]
y_est = momentos_img[1][1]/momentos_img[1][0]

#Printar e desenhar os centróides calculados (através das relaçõs m10/m00 e m01/m00)

print("Centroide triângulo = (", x_tri, ", ",y_tri,")")
print("Centroide estrela = (", x_est, ", ",x_est,")")
print("\n")

#Resultados
plt.figure(figsize = (15,15))
plt.subplot(2,2,1)
plt.imshow(img_tri_g,cmap = "gray")
plt.title("Imagem Triângulo (Escala de Cinza) - Não Rotacionado")

plt.subplot(2,2,2)
plt.imshow(img_est_g,cmap = "gray")
plt.title("Imagem Estrela Original (Escala de Cinza) - Não Rotacionado")

plt.subplot(2,2,3)
plt.imshow(img_tri)
plt.plot(x_tri,y_tri,"+")
plt.title("Imagem Triângulo com Centróide - Não Rotacionado")

plt.subplot(2,2,4)
plt.imshow(img_est)
plt.plot(x_est,y_est,"+")
plt.title("Imagem Estrela com Centróide - Não Rotacionado")

# Rotacionado as Imagems em 180

img_tri_rot = cv.rotate(img_tri, cv.ROTATE_180)
img_est_rot = cv.rotate(img_est, cv.ROTATE_180)

#Converter imagens do triângulo e da estrela em gray scale
img_tri_rot_g = cv.cvtColor(img_tri_rot, cv.COLOR_BGR2GRAY)
img_est_rot_g = cv.cvtColor(img_est_rot, cv.COLOR_BGR2GRAY)

#Momentos m00, m10, m01 e m11 das imagens
momentos_img_rot = []
momentos_img_rot.append(momentos(img_tri_rot_g))
momentos_img_rot.append(momentos(img_est_rot_g))

print("Momentos - Imagem Triângulo Rotacionado :")
print("  Momento 00: {}".format(momentos_img_rot[0][0]))
print("  Momento 01: {}".format(momentos_img_rot[0][1]))
print("  Momento 10: {}".format(momentos_img_rot[0][2]))
print("  Momento 11: {}".format(momentos_img_rot[0][3]))

print("\n")

print("Momentos - Imagem Estrela Rotacionado :")
print("  Momento 00: {}".format(momentos_img_rot[1][0]))
print("  Momento 01: {}".format(momentos_img_rot[1][1]))
print("  Momento 10: {}".format(momentos_img_rot[1][2]))
print("  Momento 11: {}".format(momentos_img_rot[1][3]))
print("\n")


# Centro de Massas das Imagens

x_tri = momentos_img_rot[0][2]/momentos_img_rot[0][0]
y_tri = momentos_img_rot[0][1]/momentos_img_rot[0][0]
x_est = momentos_img_rot[1][2]/momentos_img_rot[1][0]
y_est = momentos_img_rot[1][1]/momentos_img_rot[1][0]

print("Centroide Triângulo Rotacionado = (", x_tri, ", ",y_tri,")")
print("Centroide Estrela Rotacionado = (", x_est, ", ",x_est,")")

print("\n")

#Resultados
plt.figure(figsize = (15,15))
plt.subplot(2,2,1)
plt.imshow(img_tri_rot_g,cmap = "gray")
plt.title("Imagem Triângulo (Escala de Cinza) - Rotacionada")

plt.subplot(2,2,2)
plt.imshow(img_est_rot_g,cmap = "gray")
plt.title("Imagem Estrela Original (Escala de Cinza) - Rotacionada")

plt.subplot(2,2,3)
plt.imshow(img_tri_rot)
plt.plot(x_tri,y_tri,"+")
plt.title("Imagem Triângulo com Centróide - Rotacionada")

plt.subplot(2,2,4)
plt.imshow(img_est_rot)
plt.plot(x_est,y_est,"+")
plt.title("Imagem Estrela com Centróide - Rotacionada")
````

````python
def assinatura(img, ic, jc):
  '''
  Descrição: Cria assinatura para uma imagem.

  Entrada:
    - img: Imagem de entrada (uint8).
    - ic: Coordenada X do centro de massa normalizado (uint8).
    - jc: Coordenada Y do centro de massa normalizado (uint8).
  
  Saída:
    - rho: Coordenadas polar (rho) dos pontos [range=0:img.shape[0]*np.sqrt(2)/2].
    - theta: Coordenadas polar (theta) dos pontos [range=-180:180].
  '''
  # Sistema de coordenadas
  # .__> jj (y)
  # |   
  # v
  # ii (x)

  # Teste de insanidade
  if img.ndim != 2:
    raise ValueError('Imagem deve ter 2 dimensões.')

  # Varre a imagem, determinando as coordenadas dos pontos brancos
  ii,jj = np.nonzero(img)

  # Translada a origem do sistema de coordenadas para o centro de massa
  ii -= int(ic)
  jj -= int(jc)

  # Converte as novas coordenadas cartesianas para coordenadas polares
  rho = np.sqrt(ii**2 + jj**2)
  theta = np.rad2deg(np.arctan2(jj, ii))

  return rho, theta


## -- Seu código começa AQUI -- ##
#Importar as imagens em grayscale
img_1 = cv.imread("circulo1.tif",0)
img_2 = cv.imread("circulo2.tif",0)
img_3 = cv.imread("quadrado1.tif",0)
img_4 = cv.imread("quadrado2.tif",0)

#Função para o cálculo dos momentos
(momento_00_1,momento_10_1,momento_01_1, momento_11_1) = momentos(img_1)
(momento_00_2,momento_10_2,momento_01_2, momento_11_2) = momentos(img_2)
(momento_00_3,momento_10_3,momento_01_3, momento_11_3) = momentos(img_3)
(momento_00_4,momento_10_4,momento_01_4, momento_11_4) = momentos(img_4)

#Cálculo dos centróides (xi,yi)
x1, y1 = np.uint8(momento_10_1/momento_00_1), np.uint8(momento_01_1/momento_00_1)

x2, y2 = np.uint8(momento_10_2/momento_00_2), np.uint8(momento_01_2/momento_00_2)

x3, y3 = np.uint8(momento_10_3/momento_00_3),np.uint8(momento_01_3/momento_00_3)

x4, y4 = np.uint8(momento_10_4/momento_00_4), np.uint8(momento_01_4/momento_00_4)

#Aplicação da função assinatura
rho1, theta1 = assinatura(img_1,x1,y1)
rho2, theta2 = assinatura(img_2,x2,y2)
rho3, theta3 = assinatura(img_3,x3,y3)
rho4, theta4 = assinatura(img_4,x4,y4)

#Resultados
plt.figure(figsize = (15,15))
plt.subplot(4,2,1)
plt.imshow(img_1,cmap = 'gray')
plt.plot(x1,y1,'+')
plt.title("Circulo 1")

plt.subplot(4,2,2)
plt.plot(theta1,rho1,'.')
plt.title("Assinatura do Circulo 1")

plt.subplot(4,2,3)
plt.imshow(img_2,cmap = 'gray')
plt.plot(x2,y2,'+')
plt.title("Circulo 2")

plt.subplot(4,2,4)
plt.plot(theta2,rho2,'.')
plt.title("Assinatura do Circulo 2")

plt.subplot(4,2,5)
plt.imshow(img_3,cmap = 'gray')
plt.plot(x3,y3,'+')
plt.title("Quadrado 1")

plt.subplot(4,2,6)
plt.plot(theta3,rho3,'.')
plt.title("Assinatura do Quadrado 1")

plt.subplot(4,2,7)
plt.imshow(img_4,cmap = 'gray')
plt.plot(x4,y4,'+')
plt.title("Quadrado 2")

plt.subplot(4,2,8)
plt.plot(theta4,rho4,'.')
plt.title("Assinatura do Quadrado 2")

plt.show()
````

#### Comentários:

Por meio  da função "momentos", elaborada no primeiro exercícios, encontramos as coordenadas dos centróides da imagens dos círculos e dos quadrados, de mod que obtemos as assinaturas dos contorno das imagens. 

A construção das assinaturas consiste em duas principais etapas: 

- Translação da origem do sistema de coordenadas dos pixels para o centróide das figuras;
- Conversão das coordenadas cartesianas em polares, visando uma varredura radial ao redor da imagem; 

A partir das assinaturas geradas, percebemos a variação da aparência e a periodicidade dos pontos plotados, isso é de acordo com a posição do centróide das imagens, assim como de acordo com o formato das figuras e a grossura dos contornos.

````python
#Importando imagens originais
img_d = cv.imread('folha_ferrugem.png')
img_s = cv.imread('folha_saudavel.png')

#Converter BGR to RGB
img_d = cv.cvtColor(img_d,cv.COLOR_BGR2RGB)
img_s = cv.cvtColor(img_s,cv.COLOR_BGR2RGB)

#Extração dos canais R, G e B da imagem
(R_d,G_d,B_d) = cv.split(img_d) # Canais RGB da imagem da folha doente
(R_s,G_s,B_s) = cv.split(img_s) # Canais RGB da imagem da folha saldável

#Matrizes para armazenar as máscaras ds canais R, G e B
MR_d = np.zeros([img_d.shape[0],img_d.shape[1]],dtype = np.uint8)  #mascara para o canal R - Folha Doente
MG_d = np.zeros([img_d.shape[0],img_d.shape[1]],dtype = np.uint8)  #mascara para o canal B - Folha Doente
MB_d = np.zeros([img_d.shape[0],img_d.shape[1]],dtype = np.uint8)  #mascara para o canal G - Folha Doente

MR_s = np.zeros([img_s.shape[0],img_s.shape[1]],dtype = np.uint8)  #mascara para o canal R - Folha Doente
MG_s = np.zeros([img_s.shape[0],img_s.shape[1]],dtype = np.uint8)  #mascara para o canal B - Folha Doente
MB_s = np.zeros([img_s.shape[0],img_s.shape[1]],dtype = np.uint8)  #mascara para o canal G - Folha Doente


# Realizando binarização dos canais R, G e B

for row in range (img_d.shape[0]):
  for col in range(img_d.shape[1]):
    if R_d[row,col] >= 200:
      MR_d[row,col] = 255
      MR_s[row,col] = 255

    if G_d[row,col] >= 200:
      MG_d[row,col] = 255
      MG_s[row,col] = 255

    if B_d[row,col] <= 100:
      MB_d[row,col] = 255 
      MB_s[row,col] = 255 


# Mascara Resultando - Multiplicação das máscaras de R, G e B - Identificação da Ferrugem na Folha
Mascara_R_d = MR_d*MG_d*MB_d  # Mascara da Folha Doente
Mascara_R_s = MR_s*MG_s*MB_s  # Mascara da Folha Saldável

''' Descritor de cor ''' 

# Converter imagem original para grayscale
img_d_gray = cv.cvtColor(img_d, cv.COLOR_RGB2GRAY)     
img_s_gray = cv.cvtColor(img_s, cv.COLOR_RGB2GRAY)        

# Binarizar imagem da planta para criar uma masácara inteira dela

ret_d,mp_d = cv.threshold(img_d_gray,0,255,cv.THRESH_BINARY_INV+cv.THRESH_OTSU)  
ret_s,mp_s = cv.threshold(img_s_gray,0,255,cv.THRESH_BINARY_INV+cv.THRESH_OTSU) 

# Complemento da máscara
mp_d_comp = cv.bitwise_not(mp_d)       
mp_s_comp = cv.bitwise_not(mp_s)   

#Variável que armazena o número de pixels total da folha 
#Variável que armazena o número de pixels da ferrugem da folha

Pfl_d, Pfr_d = 0,0    # Folha Doente 
Pfl_s, Pfr_s = 0,0    # Folha Saldável
         
for row in range (img_d.shape[0]):
  for col in range(img_d.shape[1]):
    if(mp_d_comp[row,col] > 0):
      Pfl_d = Pfl_d + 1
    if(mp_s_comp[row,col] > 0):
      Pfl_s = Pfl_s + 1      
    if(Mascara_R_d[row,col] > 0):
      Pfr_d = Pfr_d + 1
    if(Mascara_R_s[row,col] > 0):
      Pfr_s = Pfr_s + 1
      
print("Folha Defeituosa: Cory(%) = ",(Pfr_s/Pfl_s) * 100)
print("Folha Saldável : Cory(%) = ",(Pfr_d/Pfl_d) * 100)


#Resultados
plt.figure(figsize = (15,8))
plt.subplot(3,4,1)
plt.imshow(img_d)
plt.title("Folha Doente - Imagem original")

plt.subplot(3,4,2)
plt.imshow(R_d, cmap = 'gray')
plt.title("Folha Doente - Canal R")

plt.subplot(3,4,3)
plt.imshow(G_d, cmap = 'gray')
plt.title("Folha Doente - Canal G")

plt.subplot(3,4,4)
plt.imshow(B_d, cmap = 'gray')
plt.title("Folha Doente - Canal B")


plt.subplot(3,4,5)
plt.imshow(MR_d, cmap = 'gray')
plt.title("Folha Doente - Máscara canal R")


plt.subplot(3,4,6)
plt.imshow(MG_d, cmap = 'gray')
plt.title("Folha Doente - Máscara canal G")

plt.subplot(3,4,7)
plt.imshow(MB_d, cmap = 'gray')
plt.title("Folha Doente - Máscara canal B")

plt.subplot(3,4,8)
plt.imshow(Mascara_R_d, cmap = 'gray')
plt.title("Folha Doente - Máscara resultante")

plt.subplot(3,4,9)
plt.imshow(mp_d_comp, cmap = 'gray')
plt.title("Folha Doente - Máscara da planta")

#Resultados
plt.figure(figsize = (15,8))
plt.subplot(3,4,1)
plt.imshow(img_s)
plt.title("Folha Saldável - Imagem original")

plt.subplot(3,4,2)
plt.imshow(R_s, cmap = 'gray')
plt.title("Folha Saldável - Canal R")

plt.subplot(3,4,3)
plt.imshow(G_s, cmap = 'gray')
plt.title("Folha Saldável - Canal G")

plt.subplot(3,4,4)
plt.imshow(B_s, cmap = 'gray')
plt.title("Folha Saldável - Canal B")


plt.subplot(3,4,5)
plt.imshow(MR_s, cmap = 'gray')
plt.title("Folha Saldável - Máscara canal R")


plt.subplot(3,4,6)
plt.imshow(MG_s, cmap = 'gray')
plt.title("Folha Saldável - Máscara canal G")

plt.subplot(3,4,7)
plt.imshow(MB_s, cmap = 'gray')
plt.title("Folha Saldável - Máscara canal B")

plt.subplot(3,4,8)
plt.imshow(Mascara_R_s, cmap = 'gray')
plt.title("Folha Saldável - Máscara resultante")

plt.subplot(3,4,9)
plt.imshow(mp_s_comp, cmap = 'gray')
plt.title("Folha Saldável - Máscara da planta")
````
