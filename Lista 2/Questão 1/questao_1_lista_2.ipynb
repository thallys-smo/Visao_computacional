```python
# Importanto Bibliotecas 

import cv2 as cv
import numpy as np
from matplotlib import pyplot as plt
```

1. Leia a imagem bird.png, mostre seu formato e verifique o número de canais. Decomponha a imagem lida em suas três componentes RGB. Mostre a imagem original e suas componentes em um subplot de 4 colunas. Identifique cada imagem colocando um título em cada uma. Comente os resultados.

```python
# Lendo Imagem: bird.png

local_img = '/home/thallys/Documentos/Graduação/Visão Computacional/Visao_computacional/Lista 2/Arquivos/bird.png'

#leitura de imagem
img = cv.imread(local_img) 

# pega as dimensões da imagem
dimensao = img.shape 

# Altura, largura e canais da imagem
altura  = img.shape[0]
largura = img.shape[1]
canais  = img.shape[2]

print('Dimensão da imagem         : ',dimensao)
print('Altura da imagem           : ',altura)
print('Largura da imagem          : ',largura)
print('Número de canais da imagem : ',canais)

# Mostra imagem original

plt.imshow(img)
plt.title('Original - {} x {}'.format(altura, largura))
plt.show()

# Decompondo a imagem no canais em seus canais

B, G, R = cv.split(img)

plt.title("Canal R")
plt.imshow(R)
plt.show()

plt.title("Canal G")
plt.imshow(G)
plt.show()

plt.title("Canal B")
plt.imshow(B)
plt.show()

```

```python
# Plotando todas as imagens juntas
plt.figure(figsize=(20,60))
plt.subplot(1,4,1)
plt.title("Original")
plt.imshow(img)

plt.subplot(1,4,2)
plt.title("Canal Red")
plt.imshow(R)

plt.subplot(1,4,3)
plt.title("Canal Green")
plt.imshow(G)

plt.subplot(1,4,4)
plt.title("Canal Blue")
plt.imshow(B)
plt.show()
```

2. Transforme-a para HSV. Mostre a imagem original e suas componentes em um subplot de 4 colunas. Identifique cada imagem colocando um título em cada uma. Comente os resultados.

```python
img_HSV	= cv.cvtColor(img, cv.COLOR_BGR2HSV)
H, S, V = cv.split(img_HSV)

plt.title("Canal H")
plt.imshow(H)
plt.show()

plt.title("Canal S")
plt.imshow(S)
plt.show()

plt.title("Canal V")
plt.imshow(B)
plt.show()

```

```python
# Plotando todas as imagens juntas
plt.figure(figsize=(20,60))
plt.subplot(1,4,1)
plt.title("Original")
plt.imshow(img_HSV)

plt.subplot(1,4,2)
plt.title("Matriz")
plt.imshow(H)

plt.subplot(1,4,3)
plt.title("Saturação")
plt.imshow(S)

plt.subplot(1,4,4)
plt.title("Valor")
plt.imshow(V)
plt.show()
```
