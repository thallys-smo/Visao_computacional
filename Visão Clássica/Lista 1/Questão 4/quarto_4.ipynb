```python
import cv2 as cv
import numpy as np
from matplotlib import pyplot as plt
```

```python
'''Leitura e recolhimento de dados da imagem'''

# Local da Imagem
local_img = "/home/thallys/Documentos/Graduao/Viso Computacional/Visao_computacional/Lista 1/Arquivos/board_ruido.tif"

# Leitura de imagem
img = plt.imread(local_img, 0) 

# Kernel
kernel_3x3 = np.ones((3,3)) / 9
print(kernel_3x3)

img_3x3 = cv.filter2D(src=img, ddepth=-1, kernel=kernel_3x3)

kernel_5x5 = np.ones((5,5)) / 9
print(kernel_5x5)

img_5x5 = cv.filter2D(src=img, ddepth=-1, kernel=kernel_5x5)

plt.imshow(img)
plt.title('Origianl - Com Rudo')
plt.show()

plt.imshow(img_3x3)
plt.title('Alterada - 3x3')
plt.show()

plt.imshow(img_5x5)
plt.title('Alterada - 5x5')
plt.show()
```
