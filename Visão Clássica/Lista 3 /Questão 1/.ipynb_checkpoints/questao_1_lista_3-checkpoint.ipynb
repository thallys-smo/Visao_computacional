```python
import numpy as np
import matplotlib.pyplot as plt
import cv2 as cv
```

#### 1. Operações lógicas em imagens binárias

Faça a leitura das imagens A.png e B.png, transforme-as para o tipo binário (logical) e realize as operações abaixo. Mostre as imagens originais e todos os resultados utilizando subplots. Coloque título em cada uma delas. Comente os resultados encontrados.

União, Interseção, Diferença e Diferença 

```python
# Caminhos das Imagens A.png e B.png
local_img_A = "/home/thallys/Documentos/Graduação/Visão Computacional/Visao_computacional/Lista 3 /Arquivos/A.png"
local_img_B = "/home/thallys/Documentos/Graduação/Visão Computacional/Visao_computacional/Lista 3 /Arquivos/B.png"

# Lendo as imagens acima
img_A = cv.imread(local_img_A)
img_B = cv.imread(local_img_B)

# Plotando Imagens
plt.figure(figsize=(15, 7)) 

plt.subplot(2,3,1)
plt.imshow(img_A)
plt.title('Imagem A')

plt.subplot(2,3,2)
plt.imshow(img_B)
plt.title('Imagem B')

plt.show()
```

```python
# Operações com as Imagens
uniao_AB = np.bitwise_or(img_A,img_B)   # 'OR' União
inter_AB = np.bitwise_and(img_A, img_B) # 'AND' Interseção
dif_AB = np.bitwise_and(img_A,~img_B)   # Diferença A-B
dif_BA = np.bitwise_and(img_B,~img_A)   # Diferença B-A
```

```python
# Plotando todas as imagens juntas

plt.figure(figsize = (12,12))
plt.subplot(3,2,1)
plt.title("Imagem A - Original")
plt.imshow(img_A)

plt.subplot(3,2,2)
plt.imshow(img_B)
plt.title("Imagem B - Original")

plt.subplot(3,2,3)
plt.imshow(uniao_AB)
plt.title("União AB")

plt.subplot(3,2,4)
plt.imshow(inter_AB)
plt.title("Interseção AB")

plt.subplot(3,2,5)
plt.imshow(dif_AB)
plt.title("Diferença AB")

plt.subplot(3,2,6)
plt.imshow(dif_BA)
plt.title("Diferença BA")

plt.show()
```

### Comentários: 

Na operação União, observamos que ambos os elementos de A e de B foram mantidos, resultando em uma figura completa do sorriso. Enquanto na operação Interseção, somente os pixels comuns das imagens A e B foram mantido, de modo que os pixels próprios de cada imagens separadas foram eliminados. 
Continuando, analisando as operações de subtração, observamos que  eliminamos na imagem resultante parcelas de A ou de B, ou seja, apenas mostramos os elementos próprios de A, quando analisamos o caso A-B, mas no caso B-A, somente os elementos próprios de B.
