```python
import numpy as np
import matplotlib.pyplot as plt
import cv2 as cv
```

```python
# Local da imagem shapes.png
local_img = "/home/thallys/Documentos/Graduação/Visão Computacional/Visao_computacional/Lista 3 /Arquivos/shapes.png"
local_img_resul = "/home/thallys/Documentos/Graduação/Visão Computacional/Visao_computacional/Lista 3 /Arquivos/shapes_resultado.png"

# Lendo a imagem shapes.png
img_shapes = cv.imread(local_img)
img_shapes_resul = cv.imread(local_img_resul)

# Plotando a imagem shapes.png
plt.figure(figsize=(15, 7)) 

plt.subplot(2,3,1)
plt.imshow(img_shapes, cmap='gray')
plt.title('Imagem Shapes - Sem Processamento')

plt.subplot(2,3,2)
plt.imshow(img_shapes_resul, cmap='gray')
plt.title('Imagem Shapes - Com Processamento')

plt.show()
```

```python
#Elemento estruturante quadrado de largura 10
SE = np.ones((9,9),np.uint8)

# Realizando uma Abertura
abertura_9 = cv.morphologyEx(img_shapes, cv.MORPH_OPEN , SE)

# Realizando um Fechamento
fechamento_9 = cv.morphologyEx(img_shapes, cv.MORPH_CLOSE , SE)
```

```python
# Plotando a Avertura e o Fechamento

plt.figure(figsize=(15, 7)) 

plt.subplot(2,3,1)
plt.imshow(abertura_9, cmap='gray')
plt.title('Imagem Shapes - Abertura')

plt.subplot(2,3,2)
plt.imshow(fechamento_9, cmap='gray')
plt.title('Imagem Shapes - Fechamento')

plt.show()
```

```python
# Subtração das Imagens: Verificação

img_aber_fecha = np.bitwise_and(abertura_9,fechamento_9)

img_resul = np.bitwise_and(img_shapes,~abertura_9)
```

```python
# Plotando o Resultado Final

plt.figure(figsize=(15, 7)) 

plt.subplot(2,3,1)
plt.imshow(img_shapes_resul, cmap='gray')
plt.title('Imagem Shapes Resultando Fornececida')

plt.subplot(2,3,2)
plt.imshow(img_resul, cmap='gray')
plt.title('Imagem Shapes Resultando Elaborada')

plt.show()
```
