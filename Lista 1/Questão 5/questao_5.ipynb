```python
import cv2 as cv
import numpy as np
from matplotlib import pyplot as plt
```

```python
'''Leitura e recolhimento de dados da imagem'''

# Local da Imagem
local_img = "/home/thallys/Documentos/Graduação/Visão Computacional/Visao_computacional/Lista 1/Arquivos/circuit.tif"

# Leitura de imagem
img = plt.imread(local_img, 0) 

# Mostrando a imagem analisada
plt.imshow(img, cmap='gray')
plt.title('Original')
plt.show()
```

```python
''' Filtragem e Impressão das Imagens'''
# Kernel: filtro espacial passa alta

kernel = np.array(((1,  1,   1),
                   (1, -8,   1),
                   (1,  1,  1))) / 8

print("Kernel:",kernel)

img_filtrada = cv.filter2D(src=img, ddepth=-1, kernel=kernel)

plt.imshow(img, cmap='gray')
plt.title('Original')
plt.show()

plt.imshow(img_filtrada, cmap='gray')
plt.title('Alterada - Filtro Passa Alta')
plt.show()

plt.imshow(255 - img_filtrada, cmap = 'gray')
plt.title('Alterada - Filtro Passa Alta Negativado')
plt.show()
```

```python
# Plotando todas as imagens juntas

plt.figure(figsize=(15, 7)) 

plt.subplot(2,3,1)
plt.imshow(img,cmap='gray')
plt.title('Original')

plt.subplot(2,3,2)
plt.imshow(img_filtrada, cmap='gray')
plt.title('Alterada - Filtro Passa Alta')

plt.subplot(2,3,3)
plt.imshow(255 - img_filtrada, cmap = 'gray')
plt.title('Alterada - Filtro Passa Alta Negativado')

plt.show()
```

```python
#Kernel para filtro de aguçamento. É igual a soma do template de um filtro passa alta com o template identidade normalizado (mesma normalização que o template passa-alta) 
kernel_agu = kernel + np.array(((0,0,0),
                               (0,9,0),
                               (0,0,0)))/9

img_agucado = cv.filter2D(img,-1,kernel_agu)

plt.imshow(img_agucado, cmap='gray')
plt.title('Alterada - Filtro Aguçado')
plt.show()
```

```python
# Plotando todas as imagens juntas

plt.figure(figsize=(15, 7)) 

plt.subplot(2,3,1)
plt.imshow(img,cmap='gray')
plt.title('Original')

plt.subplot(2,3,2)
plt.imshow(img_filtrada, cmap='gray')
plt.title('Alterada - Filtro Passa Alta')

plt.subplot(2,3,3)
plt.imshow(255 - img_filtrada, cmap = 'gray')
plt.title('Alterada - Filtro Passa Alta Negativado')

plt.subplot(2,3,4)
plt.imshow(img_agucado, cmap='gray')
plt.title('Alterada - Filtro Aguçado')

plt.show()
```
